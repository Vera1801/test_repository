```python
#задание 1
log=pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
def filtered(row):
    if row['traffic_source']=='yandex':
        return 'organic'
    elif row['traffic_source']=='google':
        return 'organic'
    elif row['traffic_source']=='paid':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    elif row['traffic_source']=='email':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
```

```python
log['source_type']=log.apply(filtered, axis=1)
```

```python
t=ratings.groupby('userId').agg({'timestamp':['count','min','max']}).reset_index()
t.columns=['userId','timestamp_count','timestamp_min','timestamp_max']
t=t[t.timestamp_count>100].reset_index(drop=True)
t['timestamp_diff']=t['timestamp_max']-t['timestamp_min']
t.timestamp_diff.mean()
```
