Исходные данные имеют следующую структуру:

1. перечень всех документов

documents = [
{‘type’: ‘passport’, ‘number’: ‘2207 876234’, ‘name’: ‘Василий Гупкин’},
{‘type’: ‘invoice’, ‘number’: ‘11-2’, ‘name’: ‘Геннадий Покемонов’},
{‘type’: ‘insurance’, ‘number’: ‘10006’, ‘name’: ‘Аристарх Павлов’}
]

2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

directories = {
‘1’: [‘2207 876234’, ‘11-2’],
‘2’: [‘10006’],
‘3’: []
}

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды “q”).

```python
documents = [
{"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
{"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
{"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
"1": ["2207 876234", "11-2"],
"2": ["10006"],
"3": []
}
```

```python
def find_data(doc_N,meta_d):
#doc_N=input("Введите номер документа \n")    
    for doc in documents:
        if(doc["number"]==doc_N):
            result="Владелец документа: \n"+doc[meta_d]
            break
        else: 
            result="Документ не найден в базе \n"
    return result

def find_rack(doc_N):
#doc_N=input("Введите номер документа \n")    
    for rack, list_doc in directories.items():
        for doc in list_doc: 
            if doc==doc_N:
                result=rack
                break
    return result

def add_rack(rack_N):
    if rack_N in directories:
        racks_str = ",".join(list(directories.keys()))
        result="Такая полка уже существует. Текущий перечень полок: "+racks_str
    else:
        directories[rack_N]=[]
        racks_str = ",".join(list(directories.keys()))
        result="Полка добавлена. Текущий перечень полок: "+racks_str
    return result    

def del_rack(rack_N):
    if rack_N in directories:
        if len(directories[rack_N])==0:
            del directories[rack_N]
            racks_str = ",".join(list(directories.keys()))
            result="Полка удалена. Текущий перечень полок: "+racks_str
        else:
            racks_str=",".join(list(directories.keys()))
            result="На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: "+racks_str
    else:
        racks_str = ",".join(list(directories.keys()))
        result="Такой полки не существует. Текущий перечень полок: "+racks_str
    return result 

new_input=""
while (new_input!="q"):
    new_input=input("Введите команду: \n")
    if(new_input=="p"):
        print(find_data(input("Введите номер документа \n"),"name"))
    elif(new_input=="s"):
        try:
            print("Документ хранится на полке: \n"+find_rack(input("Введите номер документа \n")))
        except UnboundLocalError:
            print("Документ не найден в базе \n")
    elif(new_input=="q"):
        print("Программа завершена \n")
    elif(new_input=="l"):  
        for doc in documents: 
            print("№: "+doc["number"]+", тип: "+doc["type"]+", владелец: "+doc["name"]+
                  ", полка хранения: "+find_rack(doc["number"]))
    elif(new_input=="as"):
        print(add_rack(input("Введите номер полки \n")))
    elif(new_input=="ds"):
        print(del_rack(input("Введите номер полки \n")))
    else:
        print("Такой команды нет \n")  
```
