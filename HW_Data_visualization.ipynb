```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def make_cohort(num):
    if num <=20:
        return 'Top 20'
    elif num <50:
        return '21 - 50'
    elif num <100:
        return '51 - 100'
    else:
        return '100+'
```

```python
df = pd.read_csv('2019.csv')
```

```python
df['Rank2']=df['Overall rank'].apply(make_cohort)
df.head()
```

```python
sns.pairplot(df,hue='Rank2')
```

Данные с сайта https://www.kaggle.com/unsdsn/world-happiness - обзор индекса счастья в странах мира 
(World Happiness Report), 2019 г.
Страны разделены на 4 группы - Top 20, 21 - 50, 51-100, 100+, см легенду.
Графики показывают связь или ее отсутствие между различными показателями.

Overall rank - индекс счастья
Score - рейтинг  
GDP per capita - ВВП на душу населения
Social support - социальная поддержка
Healthy life expectancy - ожидаемая продолжительность здоровой жизни
Freedom to make choices - свобода выбора
Generosity - щедрость (благородство)
Perception of corruption - отношение к коррупции 

```python
sns.scatterplot(data=df,x='Score',y='Overall rank',hue='Rank2')

#Чем выше рейтинг (Score) , тем ниже индекс счастья (Overall rank) в стране. 
```

```python
sns.scatterplot(data=df,x='Overall rank',y='GDP per capita',hue='Rank2')

#чем выше ВВП на душу населения, тем выше индекс счастья в стране (счастье зависит от благосостояния)
```

```python
sns.scatterplot(data=df,x='Overall rank',y='Healthy life expectancy',hue='Rank2')

#чем выше ожидаемая продолжительность жизни, тем выше индекс счастья в стране
```

```python
sns.scatterplot(data=df,x='Overall rank',y='Perceptions of corruption',hue='Rank2')

#отношение к коррупции у стран с высоким и низким индексом счастья не сильно отличается
```

```python
sns.scatterplot(data=df,x='Overall rank',y='Generosity',hue='Rank2')

#уровень щедрости в странах с высоким и низким индексом счастья не сильно отличается 
```
