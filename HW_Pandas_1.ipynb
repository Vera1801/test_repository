```python
#задание2. По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
#категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

power = pd.read_csv('power.csv')
power.head(7)
```

```python
p=power[power.quantity>=0]
p=p[(p.year>=2005)&(p.year<=2010)]
p=p[(p.category==4)|(p.category==12)|(p.category==21)]
p=p[p.country.isin(['Lithuania','Latvia','Estonia'])]
print("Суммарное потребление электроэнергии категорий 4, 12 и 21 за период с 2005 по 2010 гг странами Прибалтики ", p.quantity.sum())
```
